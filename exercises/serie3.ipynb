## E301 - Leer un DataFrame desde un CSV

Descarga un archivo CSV y cárgalo en un DataFrame. Muestra la información del DataFrame y su descripción (estadísticos principales).

```python
url = "https://raw.githubusercontent.com/codeforamerica/ohana-api/master/data/sample-csv/addresses.csv"
customers_df = pd.read_csv(url)
display(customers_df)
```

```python
print("Información del dataframe: \n", customers_df.info())
print("\nDescripción del dataframe: \n", customers_df.describe())
```

## E302 - Extraer las columnas de análisis

Dado un DataFrame, crea un nuevo DataFrame que contenga solo las columnas de análisis.

```python
analysis_columns = list(customers_df.columns)
analysis_columns.remove("id")
analysis_columns.remove("location_id")
new_df = customers_df[analysis_columns]
display(new_df)
```

## E303 - Limpieza y transformación de columnas

Dado un DataFrame, define una función que limpie y transforme los valores, considera que estos pueden ser nan.

```python
def clean_and_transform_nan_values(dataframe, column, default_value):
    dataframe[column].fillna(default_value, inplace=True)
```

```python
head = list(new_df.columns)
defaults = {
    "address_1": "24 Second Avenue",
    "address_2": "1st Floor",
    "city": "San Mateo",
    "state_province": "CA",
    "postal_code": "94401",
    "country": "US"
}

for column in head:
    clean_and_transform_nan_values(new_df, column, defaults[column])

display(new_df)
```
